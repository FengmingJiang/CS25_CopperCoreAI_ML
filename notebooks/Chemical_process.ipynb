```python
import pandas as pd
import numpy as np
```

```python
file_path = 'geochemistry.csv'  
df = pd.read_csv(file_path, encoding='latin1')

df.head()
```

```python
columns_to_keep = ['Year','SAMPLE NAME','LATITUDE (MIN.)','LONGITUDE (MIN.)','ROCK NAME','SAMPLE COMMENT','ALTERATION TYPE','SR(PPM): altern. values or methods', 'Y(PPM): altern. values or methods', 'SIO2(WT%): altern. values or methods', 'H2O(WT%)', 'CO2(WT%)', 'S(WT%)', 'SM(PPM): altern. values or methods', 'GD(PPM): altern. values or methods','AL2O3(WT%): altern. values or methods','TIO2(WT%): altern. values or methods','FE2O3(WT%): altern. values or methods','FEO(WT%): altern. values or methods','MNO(WT%): altern. values or methods','MGO(WT%): altern. values or methods','CAO(WT%): altern. values or methods','NA2O(WT%): altern. values or methods','K2O(WT%): altern. values or methods','P2O5(WT%): altern. values or methods','EU(PPM): altern. values or methods']
df_filtered = df[columns_to_keep]
df_filtered
```

```python
df_filtered.describe()
```

### 1. Delete samples related to hydrothermal alteration

```python
mask = df_filtered.apply(lambda row: row.astype(str).str.lower().str.contains('hydrothermal').any(), axis=1)
df_cleaned = df_filtered[~mask]
df_cleaned = df_cleaned.reset_index(drop=True)
print(f"original rows：{len(df_filtered)}，after deleting rows：{len(df_cleaned)}")
```

### 2. Change column names

```python
df_cleaned.columns = [col.replace(': altern. values or methods', '') for col in df_cleaned.columns]
df_cleaned.columns
```

### 2. Check H2O, CO2 and S which summation is < 3.5 wt%

```python
df_cleaned['LOI'] = df_cleaned[['H2O(WT%)', 'CO2(WT%)', 'S(WT%)']].sum(axis=1, skipna=True)
all_nan_mask = df_cleaned[['H2O(WT%)', 'CO2(WT%)', 'S(WT%)']].isna().all(axis=1)
df_cleaned.loc[all_nan_mask, 'LOI'] = pd.NA
df_cleaned['LOI'].describe()
```

```python
df_cleaned = df_cleaned[df_cleaned['LOI'].notna() & (df_cleaned['LOI'] < 3.5)]
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

### 3. Oxide analysis

```python
df_cleaned['Oxide'] = df_cleaned[['SIO2(WT%)', 'TIO2(WT%)', 'AL2O3(WT%)','FE2O3(WT%)','FEO(WT%)','MNO(WT%)','MGO(WT%)','CAO(WT%)','NA2O(WT%)','K2O(WT%)','P2O5(WT%)']].sum(axis=1, skipna=True)
all_nan_mask = df_cleaned[['SIO2(WT%)', 'TIO2(WT%)', 'AL2O3(WT%)','FE2O3(WT%)','FEO(WT%)','MNO(WT%)','MGO(WT%)','CAO(WT%)','NA2O(WT%)','K2O(WT%)','P2O5(WT%)']].isna().all(axis=1)
df_cleaned.loc[all_nan_mask, 'LOI'] = pd.NA
df_cleaned['Oxide'].describe()
```

```python
df_cleaned = df_cleaned[df_cleaned['Oxide'].notna() & (df_cleaned['Oxide'] < 101.5) & (df_cleaned['Oxide'] > 97.5)]
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

### 4. EU

```python
df_cleaned['Eu_star'] = np.sqrt(df_cleaned['SM(PPM)'] * df_cleaned['GD(PPM)'])
df_cleaned['Eu_anomaly_ratio'] = df_cleaned['EU(PPM)'] / df_cleaned['Eu_star']
df_cleaned['Eu_anomaly_ratio'].replace([np.inf, -np.inf], np.nan, inplace=True)
df_cleaned['Eu_anomaly_ratio'].describe()
```

```python
df_cleaned = df_cleaned[df_cleaned['Eu_anomaly_ratio'].notna()]
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

### 5. Al2O3

```python
df_cleaned = df_cleaned[df_cleaned['AL2O3(WT%)'].notna() & (df_cleaned['AL2O3(WT%)'] < 20)]
df_cleaned = df_cleaned.reset_index(drop=True)
df_cleaned
```

### 6. Unique location

```python
df_unique = df_cleaned.drop_duplicates(subset=['LATITUDE (MIN.)', 'LONGITUDE (MIN.)'], keep='first')
df_unique = df_unique.reset_index(drop=True)
df_unique
```

### 7. SiO2

```python
df_unique = df_unique[df_unique['SIO2(WT%)'] < 57]
df_unique = df_unique.reset_index(drop=True)
df_unique
```

### 8. Sr/Y

```python
df_unique['SR/Y'] = df_unique['SR(PPM)'] / df_unique['Y(PPM)']
df_unique['SR/Y'].replace([np.inf, -np.inf], np.nan, inplace=True)
df_unique = df_unique[df_unique['SR/Y'] < 35]
df_unique = df_unique.reset_index(drop=True)
df_unique
```
